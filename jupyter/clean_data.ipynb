```python
# Import Pandas
import pandas as pd
```

```python
# Read CSV
london_airbnb_listings = pd.read_csv(r'C:\Users\kpali\Documents\Projects\London_Airbnb_Search\listings.csv', delimiter = ',', encoding = 'utf-8')
london_airbnb_listings.head()
```

```python
# Remove all rows that contain null values
london_airbnb_listings.dropna(how = 'any', inplace = True)
# Remove text in bathrooms_text + remove any null cells + convert to float
london_airbnb_listings['bathrooms_text'] = london_airbnb_listings['bathrooms_text'].str.replace(r'[^0-9\.]+', '')
london_airbnb_listings['bathrooms_text'] = london_airbnb_listings['bathrooms_text'].replace('',pd.NA)
london_airbnb_listings = london_airbnb_listings.dropna(subset = ['bathrooms_text'])
london_airbnb_listings['bathrooms_text'] = london_airbnb_listings['bathrooms_text'].astype(float)
# Remove $ from price + convert to float
london_airbnb_listings['price'] = london_airbnb_listings['price'].astype(str)
london_airbnb_listings['price'] = london_airbnb_listings['price'].str.replace(r'[\$,]','')
london_airbnb_listings['price'] = london_airbnb_listings['price'].astype(float)
# Rename columns + capitalise first letter of each word
london_airbnb_listings.rename(columns = {'neighbourhood_cleansed':'neighbourhood', 'bathrooms_text':'bathrooms',
'review_scores_value': 'rating', 'room_type':'room type','minimum_nights': 'minimum nights',
'maximum_nights': 'maximum_nights', 'host_reponse_time': 'host response time', 'listing_url': 'listing url'},inplace = True)
london_airbnb_listings.columns = london_airbnb_listings.columns.str.title()
# Check null values have been removed
print((london_airbnb_listings.isnull().any(axis = 1)).sum())
# Preview Columns
london_airbnb_listings.head()
```

```python
london_airbnb_listings.to_csv('../listings.csv',index = False)
```
